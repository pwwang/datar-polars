## fct_expand

```python
fct = factor(sample(letters[:3], 20, replace = TRUE))
fct
levels(fct)
```

```python
fct = fct_expand(fct, "d", "e", "f")
fct
levels(fct)
```

## fct_explicit_na

```python
f1 = factor(c("a", "a", NA, NA, "a", "b", NA, "c", "a", "c", "b"))
fct_count(f1)
```

```python
f2 = fct_explicit_na(f1)
fct_count(f2)
```

## fct_drop

```python
fct = factor(c("a", "b"), levels = c("a", "b", "c"))
fct
levels(fct)
```

```python
ft = fct_drop(fct)
ft
levels(ft)
```

```python
ft = fct_drop(fct, only = "a")
ft
levels(ft)
```

```python
ft = fct_drop(fct, only = "c")
ft
levels(ft)
```

## fct_unify

```python
fs = [factor("a"), factor("b"), factor(c("a", "b"))]
fu = fct_unify(fs)
fu
[levels(f) for f in fu]
```
